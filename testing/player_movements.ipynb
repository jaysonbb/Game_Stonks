```python
import requests
import pandas as pd
```

```python
url = 'https://www.nba.com/stats/js/data/playermovement/NBA_Player_Movement.json'
raw_json = requests.get(url).json()
data = raw_json['NBA_Player_Movement']['rows']
```

```python
trans_logs = {}
for d in data:
    if d['Transaction_Type'] not in trans_logs.keys():
        trans_logs[d['Transaction_Type']] = [d]
    else:
        trans_logs[d['Transaction_Type']].append(d)
```

```python
def player_movements(file_name='test_data/player_movement.csv', void=False):
    url = 'https://www.nba.com/stats/js/data/playermovement/NBA_Player_Movement.json'
    raw_json = requests.get(url).json()
    df = pd.json_normalize(raw_json['NBA_Player_Movement']['rows'])
    df.to_csv(file_name, index=False)
    if void:
        return
    else:
        return df
```
