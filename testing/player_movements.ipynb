```python
import requests
import pandas as pd
import datetime as dt
```

```python
def get_roster_changes(to_csv=False, file_name='test_data/player_movement.csv'):
    url = 'https://www.nba.com/stats/js/data/playermovement/NBA_Player_Movement.json'
    i = 0
    while True:
        try:
            raw_json = requests.get(url).json()
            break
        except:
            if i == 10:
                raise StopIteration('Max tries to API (10) exceeded.')
            i += 1
            continue
    
    df = pd.json_normalize(raw_json['NBA_Player_Movement']['rows'])
    
    df.columns = [c.lower() for c in df.columns]
    df['transaction_date'] = df['transaction_date'].apply(lambda x: dt.datetime.strptime(x.replace('T',' '), "%Y-%m-%d %H:%M:%S").date())
    df['team_id'] = df['team_id'].astype(int)
    df['player_id'] = df['player_id'].astype(int)
    df['additional_sort'] = df['additional_sort'].astype(int)
    if to_csv:
        df.to_csv(file_name, index=False)
    
    return df
```

```python
data = get_roster_changes()
```

```python
data = (data.loc[data['transaction_description'].str.contains('received draft consideration') == False]
            .sort_values(by='transaction_date')
            .reset_index(drop=True))
```
